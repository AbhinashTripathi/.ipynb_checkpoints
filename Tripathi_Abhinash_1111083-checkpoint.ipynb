# Task 1.2: Recommendation engine


```python
# removing header from dataset and exporting heart_train csv without header
# disease_df.to_csv ("heart_train.csv", index = None, header=False) 
```

```python
# Load the data into RDD
disease_data = sc.textFile('heart_train.csv')
disease_data.cache()

# Spliting the RDD 
disease_ratings = disease_data.map(lambda l: l.split(','))

#prining the data rate
print("The data rate is: ", disease_ratings.count())
```

```python
# Transforming the disease_ratings RDD
disease_rate_final = disease_ratings.map(lambda data: Rating(int(data[0]), int(data[3]), float(data[8])))

# Spliting the data into training and testing
training_data, test_data = disease_rate_final.randomSplit([0.8, 0.2])

# building the ALS model on the training data
ALS_model = ALS.train(disease_rate_final,rank=10, iterations=10)

# Droping ratings column
testdata_no_rating = test_data.map(lambda p: (p[0], p[1])) 

# Predicting the model
prediction_model= ALS_model.predictAll(testdata_no_rating).map(lambda r: ((r[0], r[1]), r[2]))

# Print the first rows of the RDD
prediction_model.take(2)

predictionRate = disease_rate_final.map(lambda r: ((r[0], r[1]), r[2])).join(prediction_model)

print("The prediction rate for the data is: ", predictionRate.count())
```

```python
MSE = predictionRate.map(lambda s: (s[1][0] - s[1][1])**2).mean()
print("Mean Squared Error for data is: "+str(MSE))
```

# Task 1.3: Classification

```python
#Feature hashing and LabelPoint 
test, train = disease_ratings.randomSplit([0.5, 0.5], seed=12345)
# building HashingTf instance with 200 features
tf = HashingTF(numFeatures=200)
# Maping each data to one feature
data_test = tf.transform(test)
data_train = tf.transform(train)
# Label the features: 1 for test data, 0 for train data
Data_Test = test.map(lambda line: LabeledPoint(1,[ float(line[0]), float(line[2]), float(line[9])]))
Data_Train = train.map(lambda line: LabeledPoint(0,[ float(line[0]), float(line[2]), float(line[9])]))
#combining two dataset
Final_Data = Data_Test.union(Data_Train)
print("The test data is:", Data_Test.count())
print("The train data is:", Data_Train.count())
print("The final data is:", Final_Data.count())
```

```python
# Spliting the data into training and testing
train_samples,test_samples = Final_Data.randomSplit([0.8, 0.2])

# Training the model
model = LogisticRegressionWithLBFGS.train(train_samples)
# Creating  prediction label from the test data
predictions_label = model.predict(test_samples.map(lambda x: x.features))
# Combining original labels with the predicted labels
predict_labels = test_samples.map(lambda x: x.label).zip(predictions_label)

# Check the accuracy of the model on the test data
Modelaccuracy = predict_labels.filter(lambda x: x[0] == x[1]).count() / float(test_samples.count())
print("Model accuracy for the data is : {:.2f}".format(Modelaccuracy))
```

# Task 1.4: Clustering


```python
# Loading the dataset into a RDD
clusterRDD = sc.textFile('heart_train.csv')

# Spliting the RDD based on tab
Rddsplit = clusterRDD.map(lambda x: x.split(","))

# Transforming the split RDD by creating a list of integers
intSplit = Rddsplit.map(lambda x: [int(x[3]), int(x[4])])

# Counting the number of rows in RDD 
print("There are {} rows in the rdd_split_int dataset".format(intSplit.count()))
```

```python
# Training the model with clusters from 13 to 16 and compute Within Set Sum of Squared Error(WSSSE)
for cluster in range(13, 17):
    model = KMeans.train(intSplit, cluster, seed=1)
    Within_set_sum_square_error = intSplit.map(lambda square_error: square_error).reduce(lambda x, y: x + y)# Within Set Sum of Squared Error (WSSSE)
    print("The cluster {} has Within Set Sum of Squared Error {}".format(cluster, Within_set_sum_square_error))

# Training the model again with the best k 
KMeans_model = KMeans.train(intSplit, k=15, seed=1)

# Get the cluster centers
clusterCenters = KMeans_model.clusterCenters
```

```python
spark_session = SparkSession.builder.master("local").appName("HeartDiseasesPredict").config("spark_config","heart_diseases").getOrCreate()

# Converting split_int RDD into Spark DataFrame
splitInt_df = spark_session.createDataFrame(intSplit, schema=["col1", "col2"])

# Converting Spark DataFrame into Pandas DataFrame
splitIntdf_pandas = splitInt_df.toPandas()

# Converting "cluster_centers" into Pandas DataFrame
clusterCenters_pandas = pd.DataFrame(clusterCenters, columns=["col1", "col2"])

# Creating an overlaid scatter plot
plt.scatter(splitIntdf_pandas["col1"], splitIntdf_pandas["col2"],color='grey')
plt.scatter(clusterCenters_pandas["col1"], clusterCenters_pandas["col2"], color="red", marker="^")
plt.show()
print("Fig: Visualizing clusters for the heart attck i.e.blood pressure with respect to cholestrol amount")
```
